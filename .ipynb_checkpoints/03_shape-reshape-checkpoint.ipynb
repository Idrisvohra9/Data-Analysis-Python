```python
import numpy as np

# the shape of an array is the number of elements in each dimension

a = np.array(((1,2,3),(4,5,6)))

print(a.shape)
print()
# (2, 3)
# This means that the array has two dimensions and each nested array contains 3 elements. So, Shape = 3
a2 = np.array(((1,2,3),(4,5,6),(9,5,67)))

print("Dimension: \n",a2.ndim) # It is also a 2dim array with 3 arrays in it's outermost layer:
print("Shape: \n",a2.shape)
# (3, 3)
print()
# The elements containing in each nested array actually has to be uniform or else it throws an error 
# There is a way to solve the error by providing dtype="object" to the array function's parameter
# But still then it gives an unexpected result.

# If we create an array by providing ndmin as a parameter as done below it provides different result
# as I didn't expect:

a3 = np.array([1,2,3,4,5], ndmin=5)

print(a3)
# [[[[[1 2 3 4 5]]]]]
print("Dimension: \n",a3.ndim)
# 5
print("Shape: \n",a3.shape)
print()
# (1, 1, 1, 1, 5) (It means that the array is a 5th dim but it only has 5 elements at last dimension.)
# Calculation 1 * 1 * 1 * 1 * 5 = 5

# ### Reshaping an array:

# Reshaping means changing the shape of an array.

# By reshaping we can add or remove dimensions or change number of elements in each dimension.

a4 = np.asarray([1,3,5,6,7,8,9,0,2])

print("Array before reshaping (Original):\n",a4)
print(a4.size)
print(len(a4)) # Btw the number of elements an array contains can befound with len() as well as .size() function.
print()
# The array has 9 elements so, let's reshape it to contain 3 arrays each with three elements and 2dim

newa4 = a4.reshape(3,3) # Note: That it does'nt manipulate the original array, it returns a view so it  actually has
# To be assigned to another variable.

print("New array after reshape:\n", newa4)
print()
# Converting 1d array to 3d array:
# Note: We need to find the length of the array first in order to reshape it because we cannot have
# more elements or less elements left out as it will throw an error.

a5 = np.array([0,8,7,6,5,4,2,6])

print("Original 1d array:\n", a5)
print("Size: \n",a5.size)
print()
# Original 1d array:
#  [0 8 7 6 5 4 2 6]
# Size: 
#  8


newa5 = a5.reshape(2,1,4)# It means that the outermost dimension will have 2 arrays each of them containing
# 1 array and each will have 4 elements in it. So, 2 * 1 * 4 = 8 (Great Success..)
# Tip: The number of parameters(Numbers) passed in the .reshape() function is also a way we can determine the dept or
# dimensions the result array will have.

print("Reshaped to 3d array:\n",newa5)
print()
# Reshaping to Other shapes

newa5 = newa5.reshape(4,2,1)# 4 outer arrays 2 inner arrays 1 element in each one. (4 * 2 * 1 = 8)
# hence, Great success..

print("Reshaped to Other 3d shape:\n", newa5)
print()
# We can also specify an unknown dimension but only one! (This means that if we do not wish to calculate the dimension 
# numpy will do it for us)

# Converting to 2d array with unknown dimension:
newa5 = newa5.reshape(4,-1)

print("Converting to 2d array with unknown dimension:\n ",newa5)
print()
# We can also flatten a multidimensional array with .reshape() like this:

a6 = np.array([[[1,2,3],[5,6,7]]])

print("a6:\n",a6)
flattena6 = a6.reshape(-1)
print("Flatten a6:\n", flattena6)
```
